```python
######################
# 读入txt统计指定词频 #
######################

import re

# 文件名称（和该代码同一路径下）
readFile = '测试.txt'

 # 打开读写文件
with open(readFile,"r", encoding='UTF-8') as fileIn:
    
    # 关键词列表
    wordList = ['习近平', '企业', '经济', '小微企业'] # <-------在后面添加新的词语-------
    
    # 词频计数字典
    wordDict = {}
    
    # 读取txt
    txtContent = fileIn.read()
    print(txtContent, '\n-----\n')
                
    # 遍历文本统计
    for word in wordList:
        number = len(re.findall(re.compile(word), txtContent))
        wordDict[word] = number
    
    # 排序
    # x[1]是按字频排序，x[0]则是按字排序
    sortWord = sorted(wordDict.items(), key=lambda x:x[1],  reverse=True)
    
    # 输出
    print('输出词频统计：\n')
    for i in sortWord:
        print(i[0], '   ',  i[1])
```

```python
#coding:utf-8

# 无效符号
exclude_str = "，。！？、．（）【】<>《》=：+-*—“”…\"\[\]`' \n" 

# 读文件名称
num = 1 # 章节数目，从1开始

# 循环处理
for chapter in range(1,num+1):
    
    # 文件名称
    readFile = '红楼梦' + str(chapter) + '.txt'
    writeFile = '红楼梦字频' + str(chapter) + '.txt'
    print("读入 ", readFile, " 写出 ", writeFile, "\n")
    
    # 打开读写文件
    with open(readFile,"r", encoding='UTF-8') as fileIn ,open(writeFile,'w', encoding='UTF-8') as fileOut:
    
        # 重置计数
        word_list = []
        word_dict = {}
        
        # 添加每一个字到列表中
        for line in fileIn:
            for char in line:
                word_list.append(char)

        # 用字典统计每个字出现的个数       
        for char in word_list:
            if char not in exclude_str:
                if char.strip() not in word_dict: # strip去除各种空白
                    word_dict[char] = 1
                else :
                    word_dict[char] += 1
        #print(word_dict)

        # 排序
        #   x[1]是按字频排序，x[0]则是按字排序
        lstWords = sorted(word_dict.items(), key=lambda x:x[1],  reverse=True) 
        print(lstWords)
        
        # 统计有效字的总数
        sum = 0
        for i in range(0,len(lstWords)):
            sum = sum + lstWords[i][1]
        print("\n本章节总有效字数为: ", sum, "\n")
        
        # 所需的关键字 自己添加
        word56 = "之其亦或方于即"
        
        # 输出关键字的字频
        for i in range(0,len(lstWords)):
            if lstWords[i][0] in word56:
                cal = str(lstWords[i][0]) + "  " + str(lstWords[i][1]/sum) + "\n" #字符串
                print(cal) # 打印
                fileOut.write(cal) # 写出到txt
```
