# 读入

```python
import pandas as pd
import numpy as np
import math

np.set_printoptions(suppress=True, threshold = 7)

df = pd.read_excel('stock.xlsx', header=0, sheet_name='Sheet1')
df
```

# 函数

## 较好=1 较差=0

```python
# 第一层贝叶斯

def f1(data, flag):
    d = data.copy()
    if flag == 1:
        return d[ d['ROE均值'] >= 11.23]
    else: return d[ d['ROE均值'] < 11.23]

def f2(data, flag):
    d = data.copy()
    if flag == 1:
        return d[ (d['资产负债率均值'] >= 40) & (d['资产负债率均值'] <= 60) ]
    else: return d[ (d['资产负债率均值'] < 40) | (d['资产负债率均值'] > 60) ]
    
def f3(data, flag):
    d = data.copy()
    if flag == 1:
        return d[ d['信息披露'] == 0]
    else: return d[ d['信息披露'] == 1]
    
def count(data, a1, a2, a3):
    d = data.copy()
    d = f1(d, a1)
    d = f2(d, a2)
    d = f3(d, a3)
    numAll = d.shape[0]
    num1 = d[ d['资产经营'] >= 3.0 ].shape[0]
    num0 = d[ d['资产经营'] < 3.0 ].shape[0]
    return num0, num1, round(num0/numAll,3), round(num1/numAll,3)
        
def res(data):
    d = data.copy()
    print('a b c 0-坏 1-好')
    for i in [0,1]:
        for j in [0,1]:
            for k in [0,1]:
                print(i, j, k, count(d, i, j, k))
                
# 第二层贝叶斯

def f12(data, flag):
    d = data.copy()
    if flag == 1:
        return d[ d['综合估值'] != '低估' ]
    else: return d[ d['综合估值'] == '低估' ]

def f22(data, flag):
    d = data.copy()
    if flag == 1:
        return d[ d['资产经营'] >= 3.0 ]
    else: return d[ d['资产经营'] < 3.0 ]
    
def f32(data, flag):
    d = data.copy()
    if flag == 1:
        return d[ (d['收入今年比平均高'] >= 0) & (d['采购比平均高'] >= 0) ]
    else: return d[ (d['收入今年比平均高'] < 0) | (d['采购比平均高'] < 0) ]
    
def count2(data, a1, a2, a3):
    d = data.copy()
    d = f12(d, a1)
    d = f22(d, a2)
    d = f32(d, a3)
    numAll = d.shape[0]
    num1 = d[ d['综合能力'] >= 3.0 ].shape[0]
    num0 = d[ d['综合能力'] < 3.0 ].shape[0]
    return num0, num1, round(num0/numAll,3), round(num1/numAll,3)
        
def res2(data):
    d = data.copy()
    print('a b c 0-坏 1-好')
    for i in [0,1]:
        for j in [0,1]:
            for k in [0,1]:
                print(i, j, k, count2(d, i, j, k))
```

# 运行

```python
print('ROE均值 资产负债率均值 信息披露 => 资产经营')
res(df)
print('综合估值 资产经营 收入今年比平均高&采购比平均高 => 综合能力')
res2(df)
```
