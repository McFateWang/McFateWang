```python
import pandas as pd
import numpy as np
import math

np.set_printoptions(suppress=False, threshold = 7)

df = pd.read_excel('信息技术.xlsx', header=0, sheet_name='Sheet1')
df
```

```python
list_name = df['主营产品名称'].tolist()
list_flag = []

for i in list_name:
    if i.find('计算机') != -1:
        list_flag.append(1)
    else:
        list_flag.append(0)
df2 = df.copy()
df2['是否含计算机'] = list_flag

df2
```

# 输出

```python
df2.to_excel('result.xlsx',sheet_name='result')
print('finished.')
```
